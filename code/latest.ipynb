```python
import codecs
myfile = codecs.open(r"surah kosar.txt","r","UTF-8")
mytext = myfile.read()
print(mytext)

total_extracted_characters = []
mychar = "ا"
index1 = []
extracted_characters1 = []
a = -1
check = True
while(check):
    a = mytext.find(mychar, a+1)
    if a == -1:
        check = False
    else:
        if a!= len(mytext)-1:
            extracted_characters1.append(mytext[a])
            extracted_characters1.append(mytext[a+1])
            index1.append(a)
            index1.append(a+1)
total_extracted_characters.append(extracted_characters1)
print(extracted_characters1)
print(total_extracted_characters)
print(index1)
print(len(index1))

mychar = "ل"
index2 = []
extracted_characters2 = []
a = -1
check = True
while(check):
    a = mytext.find(mychar, a+1)
    if a == -1:
        check = False
    else:
        if a != len(mytext)-1:
            extracted_characters2.append(mytext[a])
            extracted_characters2.append(mytext[a+1])
            index2.append(a)
            index2.append(a+1)
total_extracted_characters.append(extracted_characters2)
print(extracted_characters2)
print(total_extracted_characters)
print(index2)
print(len(index2))
mychar = "م"
index3 = []
extracted_characters3 = []
a = -1
check = True
while(check):
    a = mytext.find(mychar, a+1)
    if a == -1:
        check = False
    else:
        if a != len(mytext)-1:
            extracted_characters3.append(mytext[a])
            extracted_characters3.append(mytext[a-1])
            index3.append(a)
            index3.append(a-1)
total_extracted_characters.append(extracted_characters3)
print(extracted_characters3)
print(total_extracted_characters)
print(index3)
print(len(index3))
mychar = "ن"
index4 = []
extracted_characters4 = []
a = -1
check = True
while(check):
    a = mytext.find(mychar, a+1)
    if a == -1:
        check = False
    else:
        if a !=0:
            extracted_characters4.append(mytext[a])
            extracted_characters4.append(mytext[a-1])
            index4.append(a)
            index4.append(a-1)           
total_extracted_characters.append(extracted_characters4)
print(extracted_characters4)
print(total_extracted_characters)
print(index4)
print(len(index4))

mychar = "و"
index5 = []
extracted_characters5 = []
a = -1
check = True
while(check):
    a = mytext.find(mychar, a+1)
    if a == -1:
        check = False
    else:
        if a != len(mytext)-1:
            extracted_characters5.append(mytext[a])
            extracted_characters5.append(mytext[a+1])
            index5.append(a)
            index5.append(a+1)
total_extracted_characters.append(extracted_characters5)
print(extracted_characters5)
print(total_extracted_characters)
print(index5)
print(len(index5))


mychar = "ي"
index6 = []
extracted_characters6 = []
a = -1
check = True
while(check):
    a = mytext.find(mychar, a+1)
    if a == -1:
        check = False
    else:
        if a !=0:
            extracted_characters6.append(mytext[a])
            extracted_characters6.append(mytext[a-1])
            index6.append(a)
            index6.append(a-1)           
total_extracted_characters.append(extracted_characters6)
print(extracted_characters6)
print(total_extracted_characters)
print(index6)
print(len(index6))


final_extracted_characters = list(filter(None, total_extracted_characters))
print(final_extracted_characters)
final_extracted_characters_to_hex = []
for x in final_extracted_characters:
    for y in x:
        final_extracted_characters_to_hex.append((format(ord(y),'d')))
print(final_extracted_characters_to_hex)


final_extracted_characters_hex_to_int = []
final_extracted_characters_hex_to_int = list(map(int,final_extracted_characters_to_hex))
print(final_extracted_characters_hex_to_int)
def xor(data):
    xor = []
    length = len(data)
    if length%2 == 0:
        for i in range(0,length,2):
            xor.append(format(data[i] ^ data[i+1], "08b"))
    else:
        for i in range(0, length-1,2):
            xor.append(format(data[i] ^ data[i+1],"08b"))
    return xor
final_xor = xor(final_extracted_characters_hex_to_int)
print(final_xor)
final_xor_in_string = ""
for i in final_xor:
    final_xor_in_string+=i
print(final_xor_in_string)
```
