```python
import cv2 
import numpy as np
from math import *
import os
import matplotlib.pyplot as plt
import sys

IMDIR='dots-destr'
outputdir='alifhu'
counter=0

# def file_sorter(element):
#     e=int(element[:-4])
#     return e
a=[]
value=1e-15

# for image in sorted(os.listdir(IMDIR),key=file_sorter):
for image in os.listdir(IMDIR):
    img =cv2.imread(os.path.join(IMDIR,image),cv2.IMREAD_GRAYSCALE)
    _,img=cv2.threshold(img,128,255,cv2.THRESH_BINARY)
    moments = cv2.moments(img)
    huMoments=[]
    huMoments = cv2.HuMoments(moments)
    
#    print(huMoments)
    
    for i in range(0,7):  
        a=huMoments[i]
        if a==0:
         #   print("Declaring smaller value to 0th moment")
            huMoments[i]=value
            huMoments[i] = -1*copysign(1.0, huMoments[i])*log10(abs(huMoments[i]))
            a_value=huMoments[i]
            cv2.imwrite(os.path.join(outputdir, "{}.png".format(counter)),img)
           # print(huMoments[i])
           # counter+=1
            
        else:
       #     print("calculating logHu")
            huMoments[i] = -1*copysign(1.0, huMoments[i])*log10(abs(huMoments[i]))
            b=np.round(huMoments[i],5)
      #      print("value of rounded huMoment",b)
            if b==10.19236 or b==10.08819:
                print("rounded huMoments",huMoments[i])
#                 counter+=1
                cv2.imwrite(os.path.join(outputdir, "{}.png".format(counter)),img)
    
    counter=counter+1
```
